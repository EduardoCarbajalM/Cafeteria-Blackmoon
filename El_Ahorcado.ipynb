```python
print("****************************************")
print("**************El Ahorcado***************")
print("****************************************")
print("Bienvenido al juego del ahorcado\n")

palabra_secreta = "fresa".upper()
palabra_secreta = palabra_secreta.strip()
letras_acertadas = ["_" for letra in palabra_secreta]
print(letras_acertadas)

# Mientras la persona no se ahorque ni acierte la palabra el juego continua
ahorco = False
acerto = False
errores = 0

while (not ahorco and not acerto):
  intento = input("\nDigita una letra: \n")
  intento = intento.strip().upper()
  if intento in palabra_secreta:
    indice = 0

    for letra in palabra_secreta:
      if intento == letra:
        letras_acertadas[indice] = letra
        #print(f"Encontre la letra `{letra}Â´ en la posicion {indice}")
      indice = indice + 1
  else:
    errores += 1
    oportunidades = len(palabra_secreta) - errores
    print(f"Te equivocaste de letra, te quedan {oportunidades} intentos")
  ahorco = errores == 6
  acerto = "_" not in letras_acertadas
  print(letras_acertadas)
  if ahorco:
    print("Perdiste el juego.")
  elif acerto:
    print("Ganaste el juego.")
print("\nFin del juego!")
```
